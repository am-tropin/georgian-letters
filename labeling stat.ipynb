```python
# part 1
df1 = pd.read_csv("labels/labels_georgian-letters-1-saturated_2023-04-24-07-13-10.csv")
df1
```

```python
# part 2
df2 = pd.read_csv("labels/labels_georgian-letters-2-middle_2023-04-24-09-43-46.csv")
df2
```

```python
df = pd.concat([df1, df2])
df.shape
```

```python
df["label_id"] = df["label_name"].apply(lambda x: int(x[:-1]))
df["geo_letter"] = df["label_name"].apply(lambda x: x[-1])
```

```python
df.groupby(["label_id"])["geo_letter"].value_counts().sort_values(ascending=False)
```
